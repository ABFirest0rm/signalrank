```python
def open_input(path: str):
    """Open plain or gzipped JSONL transparently."""
    if path.endswith(".gz"):
        return gzip.open(path, "rb")
    return open(path, "rb")
```

```python
import os

directory_path = r"E:\Coding\Projects\redrob-ranker"

for root, dirs, files in os.walk(directory_path):
    # Skip the .venv folder so os.walk doesn't even enter it
    if '.venv' in dirs:
        dirs.remove('.venv')

    print(f"Current Directory: {root}")
    
    # Loop through and print subdirectories
    for dir_name in dirs:
        print(f"  Sub-directory: {os.path.join(root, dir_name)}")
        
    # Loop through and print files
    for file_name in files:
        print(f"  File: {os.path.join(root, file_name)}")
        
    print("-" * 40)
```
